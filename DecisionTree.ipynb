```python
import pandas as pd

import numpy as np

movie = pd.DataFrame({"Popcorn":[1,1,0,0,1,1,0],
                      "Soda":[1,0,1,1,1,0,0],
                      "Age":[7,12,18, 35, 38, 50, 83],
                      "Like_movie":[0,0,1,1,1,0,0]})
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(criterion = "gini")
X = movie.drop("Like_movie", axis = 1)
y = movie["Like_movie"]

pred = dtree.fit(X,y).predict(X)
```

```python
from sklearn.tree import export_text

#Extract the Rules
rules = export_text(dtree, feature_names = list(X.columns))
print(rules)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

tree.plot_tree(dtree, feature_names = list(X.columns))
plt.tight_layout()
plt.show()
```

### Entropy vs Gini

* Entropy is 1 - Sum of pi * log2*pi
* Gini: 1 - sum of (pofyes)^2 -sum of (pofno)^2

* Entropy relies on Information where as Gini relies on probability of classes

* Entropy algo is slower than Gini because of Log calculation

* **Gini is good for most of the datasets and use cases where as Entropy is far better for Imbalanced Dataset.**
